Open the file first

```python
with open("finaldata.txt", encoding = 'utf-8') as f:
    data = f.read()
```

Split the lines where there is double gap in between the line and store them into chunks
Where element of chunks is greater than 3 only take these chunks only otherwise skip the chunks

```python
chunks = data.split("\n\n")

chunks = [c for c in chunks if len(c)>3]
```

Here parse the chunk and getting all the relevant answer by doing some manipulation here

```python
def parse_chunk(chunk):
    try:
        chunk = chunk.strip()
        sep_chunk = chunk.split('\n')
        username = sep_chunk[0]
        no_of_posts = int(sep_chunk[1].split("post")[0].replace(",",""))
        no_of_followers = float(sep_chunk[2].split("follower")[0].replace(",","").replace("K","").replace("M",""))
        if("K" in sep_chunk[2]):
            no_of_followers = int(no_of_followers*1000)
        elif("M" in sep_chunk[2]):
            no_of_followers = int(no_of_followers*1000000)
        else:
            no_of_followers = int(no_of_followers) 
  
        no_of_following = float(sep_chunk[3].split("following")[0].replace(",","").replace("K","").replace("M",""))
        if("K" in sep_chunk[3]):
            no_of_following = int(no_of_following*1000)
        elif("M" in sep_chunk[3]):
            no_of_following = int(no_of_following*1000000)
        else:
            no_of_following = int(no_of_following)

        
        name = sep_chunk[4]
        if(len(sep_chunk)> 5):
            type_of_page = sep_chunk[5]
            bio = "\n".join(sep_chunk[6:])
             
        else:
            type_of_page = "Unknown"
            bio = ""

        return {"username":username, "no_of_posts": no_of_posts, "no_of_followers":no_of_followers, "no_of_following":no_of_following,"name":name, "type_of_page":type_of_page,"bio":bio,"type_of_page":type_of_page,}
    except Exception as e:
        print(e,chunk)
   
    
  
    
    
```

Then create all_chunks dictionary to store all the parse chunks here

```python
all_chunks = []

for chunk in chunks:
 
    
    
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
  
    
```

Dump the output into data.json file

```python
import json 
s = json.dumps(all_chunks , indent =4)
with open("data.json","w") as f:
    f.write(s)
```
